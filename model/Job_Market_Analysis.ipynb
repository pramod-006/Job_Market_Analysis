```python
import spacy
import re
from spacy.matcher import Matcher
from collections import Counter
nlp = spacy.load("en_core_web_sm")

skill_keywords = [
    # Programming
    'python', 'java', 'c++', 'c#', 'go', 'typescript', 'bash', 'scala',

    # ML/DL
    'machine learning', 'deep learning', 'ai', 'ml', 'neural networks', 'supervised learning',
    'unsupervised learning', 'reinforcement learning', 'transfer learning', 'cnn','rnn',

    # Libraries/Frameworks
    'pytorch', 'tensorflow', 'scikit-learn', 'keras', 'xgboost', 'huggingface',

    # NLP / GenAI
    'nlp', 'transformers', 'bert', 'gpt', 'llm', 'chatgpt', 'diffusion models', 'rnn', 'attention mechanism',

    # Data
    'sql', 'nosql', 'mongodb', 'bigquery', 'pandas', 'numpy', 'matplotlib', 'seaborn', 'data wrangling',

    # Cloud / MLOps
    'aws', 'gcp', 'azure', 'docker', 'kubernetes', 'mlops', 'dvc', 'gitops', 'airflow', 'dagster',

    # Dev / Web
    'flask', 'fastapi', 'streamlit', 'gradio', 'api development', 'rest api', 'microservices',

    # Tools / Platforms
    'git', 'jira', 'tableau', 'power bi', 'excel', 'hugging face', 'langchain', 'vector db', 'pinecone', 'weaviate'
]


# Creating PhraseMatcher
phrase_matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
patterns = [nlp.make_doc(skill) for skill in skill_keywords]
phrase_matcher.add("SKILL", patterns)


def extract_skills(text):
    doc = nlp(text)
    matches = phrase_matcher(doc)
    skills = set()
    for match_id, start, end in matches:
        span = doc[start:end]
        skills.add(span.text.lower())
    return list(skills)

```

```python
df['extracted_skills'] = df['job_description_text'].apply(extract_skills)
df.head()
```

#Part 3: Skill Trend Detector

```python
from collections import Counter
import pandas as pd

all_skills = [skill for skills in df['extracted_skills'] for skill in skills]


skill_counts = Counter(all_skills)

skill_freq_df = pd.DataFrame(skill_counts.items(), columns=['skill', 'count'])


skill_freq_df = skill_freq_df.sort_values(by='count', ascending=False)


established_threshold = skill_freq_df['count'].quantile(0.55)
emerging_threshold = skill_freq_df['count'].quantile(0.45)

# Label classification
def classify_skill(count):
    if count >= established_threshold:
        return 'common/established'
    else:
        return 'new/emerging'

skill_freq_df['category'] = skill_freq_df['count'].apply(classify_skill)


print(skill_freq_df.head(50))
```

#TESTING OUR MODEL


```python
import spacy
from spacy.matcher import Matcher
import json

nlp = spacy.load("en_core_web_sm")


with open("skill_frequency.json", "r") as f:
    skill_frequency = json.load(f)


skill_keywords = list(skill_frequency.keys())
skill_patterns = [[{"LOWER": token.lower()} for token in skill.split()] for skill in skill_keywords]

matcher = Matcher(nlp.vocab)
matcher.add("SKILL", skill_patterns)


def detect_skill_trends(text):
    doc = nlp(text)
    matches = matcher(doc)

    found_skills = set()
    for _, start, end in matches:
        skill_text = doc[start:end].text.lower()
        found_skills.add(skill_text)

    results = []
    for skill in found_skills:
        freq_score = skill_frequency.get(skill, 0)
        category = "established" if freq_score >= 0.05 else "emerging"
        results.append({
            "skill": skill,
            "category": category,
            "trend_score": round(freq_score, 2)
        })

    return results

job_description = "Experience with PyTorch,machine learning, and diffusion models required."
result = detect_skill_trends(job_description)

from pprint import pprint
pprint(result)
```
